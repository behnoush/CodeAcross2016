```python
import folium
import geopy
from geopy.geocoders import Nominatim
```

```python
import pandas as pd
Data1 = pd.read_excel('AbandonedUrbanProperties.xlsx')
```

```python
print Data1.columns
```

```python
def visualVacants(cityCenter,numVacant):
    geolocator = Nominatim()
    count = 0
    map_1 = folium.Map(location=cityCenter,zoom_start=12, tiles='Stamen Terrain')
    point = 1 
    for rows in Data1.values:
        
        if point <= numVacant:
            temp = rows[2]+'Louisville'
            location = geolocator.geocode(temp, timeout=10)
            if location:
                point += 1
                print location
                print temp ,";",location.latitude, location.longitude
                folium.Marker([location.latitude, location.longitude],popup=str(location)).add_to(map_1)
            else:
                count +=1
    print "The number of addresses not matched:" + str(count)
    return map_1
```

```python
louisville = [38.2500,-85.7667]
map_2 = visualVacants(louisville,10)
map_2.create_map(path='map2.html')
```
